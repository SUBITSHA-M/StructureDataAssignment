# -------------Analysis for 30 days---------------------

```python
patients= train_df.groupby("Patient-Uid")
patients.head()
```

Exclude patient records from the positive class occurring within 30 days prior to taking the "TARGET_DRUG," leaving negative class patients unaffected.

# Steps:

#### Step1:iterating through the dates of the patient records and checking if he has taken the “TARGET_DRUG”.

#### Step2:If they have taken it , note down the index of the date( or drug ) taken before 30 days of the date of the “TARGET_DRUG”.


#### Step3:Now all these position indexes are stored in the list called “cutoff_index_list”

#### Step4:During the analysis we have encountered a patient, who has takent the “TARGET_DRUG” before 30 days itself like with in 6 days of the joining the clinic or atleast when the records are started on their name.

#### Step5:Finally we also note down those patients-id also to remove them. As their data is of no use and are treated as outliers.

```python
target_count=0 

cutoff_index_list=[]
index_of_target=[]

outlier_patients=0

outlier_patients_index=[]

for group_name, group_data in patients:
    for i in range(len(group_data)):

        if group_data["Incident"].iloc[i]=="TARGET_DRUG":
            index_of_target.append(i)  
            target_count+=1
            temp = i 
            while (group_data["Date"].iloc[i]-group_data["Date"].iloc[temp-1]).days < 30:
                temp-=1
                if temp==0:
                    outlier_patients+=1
                    outlier_patients_index.append(group_name)
                    break
            cutoff_index_list.append(temp)
            break
        if i == len(group_data)-1:
            cutoff_index_list.append(len(group_data))
```

```python
print("Patient count who taken TARGET_DRUG :", target_count)

print("Patients who are administered TARGET_DRUG within 30 days of given records itself :", outlier_patients)
```

# Feature Engineering

#### Time based feature engineering

```python
train_df=train_df[train_df['Patient-Uid']!='a0ee64e1-1c7c-11ec-becd-16262ee38c7f']
grouped= train_df.groupby('Patient-Uid')
print(grouped.ngroups)
```

```python
# Define pds as an iterator over your patient data
# Example data structure:
# pds = [(patient_id_1, patient_data_1), (patient_id_2, patient_data_2), ...]

count = 0
Time = []

# Iterate through the records of each patient
for patient_id, patient_data in grouped:
    # List to store the gap history of each patient
    tem = []
    # Loop through the patient history
    for j in range(train_updated_group['cutoff_indices'].iloc[count] - 1):
        # Calculate the difference between consecutive dates in days
        gap = (patient_data["Date"].iloc[j + 1] - patient_data["Date"].iloc[j]).days
        tem.append(gap)
    
    # Pad the gap vector with zeros to match the max_length
    tem.extend([0] * (max(cutoff_index_list) - len(tem)))

    # Finally, add the patient's gap history to the list
    Time.append(tem)
    
    count += 1  # Increment the count for the next patient
```

```python
print(len(Time),len(Time[0]))
```

```python
#add them to the main dataset
train_updated_group['Time_based_features']=Time
```

### ------- Model building--------

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Flatten, Dense, Dropout, concatenate
from tensorflow.keras.regularizers import l2
from tensorflow.keras.initializers import HeNormal
from keras.models import Model

# Input layers
input_layer_1 = Input(shape=pad_train.shape[1], name="Input_layer_for_LSTM_based_features")
input_layer_2 = Input(shape=(len(X_train['Time_based_features'][0]),), name='Input_layer_for_Time_based_features')
input_layer_3 = Input(shape=X_train_tfidf.shape[1], name="Input_layer_for_frequency_based_features")

# Embedding layers
embedding_layer_1 = Embedding(input_dim=57, output_dim=128, input_length=pad_train.shape[1], name="Embedding_layer_1")(input_layer_1)

# LSTM layers
lstm_layer_1 = LSTM(64, return_sequences=True, name="LSTM_layer_1")(embedding_layer_1)

# Flatten layers
flatten_layer_1 = Flatten(name="Flatten_layer_1")(lstm_layer_1)

# Dense layers
dense_0 = Dense(256, activation='relu', name='Dense_for_tfidf_input')(input_layer_3)
concatenated_tensor = concatenate([flatten_layer_1, dense_0], axis=1)
dense_1 = Dense(256, activation='relu', kernel_initializer=HeNormal(), kernel_regularizer=l2(0.0001), name='Dense_layer_after_concat')(concatenated_tensor)
drop_1 = Dropout(0.5, name="Dropout_1")(dense_1)
dense_2 = Dense(128, activation='relu', kernel_initializer=HeNormal(), kernel_regularizer=l2(0.0001), name='Dense_2')(drop_1)
drop_2 = Dropout(0.5, name="Dropout_2")(dense_2)
dense_3 = Dense(64, activation='relu', kernel_initializer=HeNormal(), kernel_regularizer=l2(0.0001), name='Dense_3')(drop_2)

# Output layer
output_layer = Dense(1, activation='sigmoid', name='Output_layer_for_binary_classification')(dense_3)

# Put all the input layers in a list format
input_layers = [input_layer_1, input_layer_2, input_layer_3]

# Define the model
model = Model(inputs=input_layers, outputs=output_layer)

# Print the model summary
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
# Specify the batch size in model.fit
batch_size = 16
model.fit(train_X, train_y, epochs=10, validation_data = (test_X,test_y),batch_size=16)
```

# F1-Score

```python
from sklearn.metrics import f1_score

predictions = model.predict(test_X)


threshold = 0.5
binary_predictions = (predictions > threshold).astype(int)


f1 = f1_score(test_y, binary_predictions)

print("F1 Score:", f1)
```
